### Task 1




##### Display the names of the unique launch sites  in the space mission


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
%sql SELECT * from SPACEXTABLE where [Launch_site] like 'CCA%' limit 5
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
%sql SELECT SUM([Payload_mass]) from SPACEXTABLE where "Customer" ='NASA (CRS)'
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
%sql SELECT AVG([Payload_mass]) from SPACEXTABLE where [Version_Booster] = 'F9 v1.1'
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
%sql SELECT "Date" from SPACEXTABLE where [Launch_outcome] like '%Success%' LIMIT 1
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
%%sql
select 'Version_Booster' from SPACEXTBL
where 'Booster_landing'='%Success%' and 'Payload_mass' between 4000 and 6000
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
%sql SELECT Count("Booster_landing") from "SPACEXTABLE" where "Booster_landing" like '%Success%'
```

```python
%%sql SELECT Count("Booster_landing") from "SPACEXTABLE" 
    where "Booster_landing" like "%Failure%"
```

### Task 8



##### List all the Version_Boosters that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
%%sql select "Version_Booster" from "SPACEXTABLE"  where 
    "Payload_mass" = (select max("Payload_mass") from "SPACEXTABLE" )
```

### Task 9


##### List the records which will display the month names, failure Booster_landings in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
%%sql SELECT 
    SUBSTR("Date", INSTR("Date", ' ') + 1, INSTR(SUBSTR("Date", INSTR("Date", ' ') + 1), ' ')) AS month,
    "Version_Booster", 
    "Launch_site" 
FROM "SPACEXTABLE" 
WHERE SUBSTR("Date", -4) = '2015' 
    AND "Booster_landing" LIKE '%Failure%' 
    AND "Booster_landing" LIKE '%drone ship%'
```

```python
%%sql select substr("Date", 4, 2) as "month", "Version_Booster", "Launch_site" 
    from "SPACEXTABLE"  where "Booster_landing" = '%Failure%' and "Date"= '%2015%'
```

```python
%sql select monthname(DATE) as month, "Booster_landing" from SPACEXTBL where DATE LIKE '2015%' and "Booster_landing" LIKE 'Failure%'
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
%%sql
select "Booster_landing", COUNT(Booster_landing) from SPACEXTBL
where Booster_landing like 'Success%' and SPACEXTBL.Date between '04/06/2010' AND '20/03/2017'
group by 1
order by 2 DESC

SELECT LandingOutcome, COUNT(Booster_landing)
        FROM SpaceX
        WHERE DATE BETWEEN '2010-06-04' AND '2017-03-20'
        GROUP BY "Booster_landing"
        ORDER BY COUNT(Booster_landing) DESC

```
